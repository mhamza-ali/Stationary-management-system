```python
import datetime
import math,random,os
import time
import tkinter
import tkinter.messagebox as tkMessageBox
from tkinter import Tk,ttk
from tkinter import *


class InventorySystem:
    def __init__(self,screen):
        self.screen = screen
        self.screen.geometry("1350x800+0+0")
        self.screen.title(" STATIONARY INVENTORY SYSTEM ")
        screen.resizable(0, 0)
        screen.config(bg="green")
        
        tops = Frame(self.screen,width=1350,height=100,bd=12,bg= "yellow",relief="raised")
        tops.pack(side="top")
        labl = Label(tops, font=("Times New Roman",28,"bold"), bg= "green" ,text = "\t\t\t\tStationary Inventory system\t\t\t\t") #main tite for the main screen
        labl.grid(row=0, column=0)
        
        
        
#==================================All variables=====================================
        
        #==================Geometry variable=================
        self.a=IntVar()
        self.scale=IntVar()
        self.c=IntVar()
        self.d=IntVar()
        
        #==================Journal variable====================
        self.h=IntVar()
        self.j=IntVar()
        self.k=IntVar()
        self.l=IntVar()
        
        #==================Painting Items variable===============
        self.z=IntVar()
        self.x=IntVar()
        self.ca=IntVar()
        self.v=IntVar()
        
        
        #==================total variable====================
        self.tg=StringVar()
        self.totals=StringVar()
        self.totald=StringVar()
        
        #==================taxes variable====================
        self.gstsa=StringVar()
        self.gsts=StringVar()
        self.gstd=StringVar()
        
        #==================calculation variable===============
        self.cname=StringVar()
        self.cphone=StringVar()
        
        self.cbill=StringVar()
        x=random.randint(1000,9999)
        self.cbill.set(str(x))
        
        self.search_bill=StringVar()
    
    
    
    
        #========frame for customer details======= F1
        F1=LabelFrame(self.screen,text = "Customer Details",font=("Times New Roman",12,"bold"),bg= "yellow",bd=10,relief="groove")
        F1.place(x=0, y=80,relwidth=1)
        
        cname_label = Label(F1, font=("Times New Roman",15,'bold'), bg= "yellow" ,text = "Customer Name").grid(row=0, column=0,pady=5)
        cname_txt = Entry(F1,textvariable=self.cname,width=18,font= "arial 15",bd=6).grid(row=0,column=1,padx=10)
        
        cphone_label = Label(F1, font=("Times New Roman",15,"bold"), bg= "yellow" ,text = "Phone number").grid(row=0, column=2,pady=5)
        cphone_txt = Entry(F1,textvariable=self.cphone,width=18,font= "arial 15",bd=6).grid(row=0,column=3,padx=10)
        
        cbill_label = Label(F1, font=("Times New Roman",15,"bold"), bg= "yellow" ,text = "Biling number").grid(row=0, column=4,pady=5)
        cbill_txt = Entry(F1,textvariable=self.search_bill,width=18,font= "arial 15",bd=6).grid(row=0,column=5,padx=10)
        
        bill_btn= Button(F1,command=self.find_bill,text="search",width=8,bd=6,font= "arial 12 bold").grid(row=0,column=6,padx=10,pady=10)
        
        
        #=======frame for Geometry========== F2
        F2=LabelFrame(self.screen,text = "All Geometry",font=("Times New Roman",15,"bold"),bg= "white",bd=10,relief="groove")
        F2.place(x=5, y=170,width=350,height=380)
        
        a_lbl = Label(F2,text="Pencil",font=("Times New Roman",16,'bold'),bg="white").grid(row=0, column=0,padx=10,pady=10,sticky="w")
        a_txt = Entry(F2,textvariable=self.a,width=10,font= ("Times New Roman",16,'bold'),bd=5,relief="sunken").grid(row=0,column=1,padx=10,pady=10)
       
        scale_lbl = Label(F2,text="Scale",font=("Times New Roman",16,'bold'),bg="white").grid(row=1, column=0,padx=10,pady=10,sticky="w")
        scale_txt = Entry(F2,textvariable=self.scale,width=10,font= ("Times New Roman",16,'bold'),bd=5,relief="sunken").grid(row=1,column=1,padx=10,pady=10)
        
        c_lbl = Label(F2,text="Eraser",font=("Times New Roman",16,'bold'),bg="white").grid(row=2, column=0,padx=10,pady=10,sticky="w")
        c_txt = Entry(F2,textvariable=self.c,width=10,font= ("Times New Roman",16,'bold'),bd=5,relief="sunken").grid(row=2,column=1,padx=10,pady=10)
        
        d_lbl = Label(F2,text="Sharpener",font=("Times New Roman",16,'bold'),bg="white").grid(row=3, column=0,padx=10,pady=10,sticky="w")
        d_txt = Entry(F2,textvariable=self.d,width=10,font= ("Times New Roman",16,'bold'),bd=5,relief="sunken").grid(row=3,column=1,padx=10,pady=10)
        
      
    
    
        
        
        #=======frame for Journal menu========== F3
        F3=LabelFrame(self.screen,text = "Journal",font=("Times New Roman",15,"bold"),bg= "white",bd=10,relief="groove")
        F3.place(x=340, y=170,width=350,height=380)
        
        h_lbl = Label(F3,text="200 Pages",font=("Times New Roman",16,'bold'),bg="white").grid(row=0, column=0,padx=10,pady=10,sticky="w")
        h_txt = Entry(F3,textvariable=self.h,width=10,font= ("Times New Roman",16,'bold'),bd=5,relief="sunken").grid(row=0,column=1,padx=10,pady=10)
       
        j_lbl = Label(F3,text="300 Pages",font=("Times New Roman",16,'bold'),bg="white").grid(row=1, column=0,padx=10,pady=10,sticky="w")
        j_txt = Entry(F3,textvariable=self.j,width=10,font= ("Times New Roman",16,'bold'),bd=5,relief="sunken").grid(row=1,column=1,padx=10,pady=10)
        
        k_lbl = Label(F3,text="400 Pages",font=("Times New Roman",16,'bold'),bg="white").grid(row=2, column=0,padx=10,pady=10,sticky="w")
        k_txt = Entry(F3,textvariable=self.k,width=10,font= ("Times New Roman",16,'bold'),bd=5,relief="sunken").grid(row=2,column=1,padx=10,pady=10)
        
        l_lbl = Label(F3,text="500 Pages",font=("Times New Roman",16,'bold'),bg="white").grid(row=3, column=0,padx=10,pady=10,sticky="w")
        l_txt = Entry(F3,textvariable=self.l,width=10,font= ("Times New Roman",16,'bold'),bd=5,relief="sunken").grid(row=3,column=1,padx=10,pady=10)
        
        
        #=======frame for menu========== F3
        F4=LabelFrame(self.screen,text = "Painting Items",font=("Times New Roman",15,"bold"),bg= "white",bd=10,relief="groove")
        F4.place(x=670, y=170,width=320,height=380)
        
        z_lbl = Label(F4,text="Oil Paint",font=("Times New Roman",16,'bold'),bg="white").grid(row=0, column=0,padx=10,pady=10,sticky="w")
        z_txt = Entry(F4,textvariable=self.z,width=10,font= ("Times New Roman",16,'bold'),bd=5,relief="sunken").grid(row=0,column=1,padx=10,pady=10)
       
        x_lbl = Label(F4,text="Oil Pastels",font=("Times New Roman",16,'bold'),bg="white").grid(row=1, column=0,padx=10,pady=10,sticky="w")
        x_txt = Entry(F4,textvariable=self.x,width=10,font= ("Times New Roman",16,'bold'),bd=5,relief="sunken").grid(row=1,column=1,padx=10,pady=10)
        
        ca_lbl = Label(F4,text="Crayons",font=("Times New Roman",16,'bold'),bg="white").grid(row=2, column=0,padx=10,pady=10,sticky="w")
        ca_txt = Entry(F4,textvariable=self.ca,width=10,font= ("Times New Roman",16,'bold'),bd=5,relief="sunken").grid(row=2,column=1,padx=10,pady=10)
        
        v_lbl = Label(F4,text="Painting Sheets",font=("Times New Roman",16,'bold'),bg="white").grid(row=3, column=0,padx=10,pady=10,sticky="w")
        v_txt = Entry(F4,textvariable=self.v,width=10,font= ("Times New Roman",16,'bold'),bd=5,relief="sunken").grid(row=3,column=1,padx=10,pady=10)
        
    
        #==========Receipt============ F5
        F5=Frame(self.screen,bd=10,relief="groove")
        F5.place(x=990, y=170,width=355,height=380)
        
        bill_title = Label(F5,text="Receipt",font="arial 15 bold",bd=6,relief="groove").pack(fill=X)
        scroll_y = Scrollbar(F5,orient=VERTICAL)
        self.txtarea = Text(F5,yscrollcommand=scroll_y.set)
        scroll_y.pack(side="right",fill=Y)
        scroll_y.config(command=self.txtarea.yview)
        self.txtarea.pack(fill="both",expand=1)
        
        
        #============frame for products total======  F6
        F6=LabelFrame(self.screen,text = "Products total",font=("Times New Roman",15,"bold"),bg= "white",bd=10,relief="groove")
        F6.place(x=0, y=560,relwidth=1,height=165)
       
        tg_lbl = Label(F6,text="Geometry Total Rs.",font=("Times New Roman",14,'bold'),bg="white").grid(row=0, column=0,padx=20,pady=1,sticky="w")
        tg_txt = Entry(F6,textvariable=self.tg,width=10,font= ("Times New Roman",16,"bold"),bd=5,relief="sunken").grid(row=0,column=1,padx=10,pady=1)

        ts_lbl = Label(F6,text="Journal Total Rs.",font=("Times New Roman",14,'bold'),bg="white").grid(row=1, column=0,padx=20,pady=1,sticky="w")
        ts_txt = Entry(F6,textvariable=self.totals,width=10,font= ("Times New Roman",16,"bold"),bd=5,relief="sunken").grid(row=1,column=1,padx=10,pady=1)
        
        td_lbl = Label(F6,text="Painting Items Total Rs.",font=("Times New Roman",14,'bold'),bg="white").grid(row=2, column=0,padx=20,pady=1,sticky="w")
        td_txt = Entry(F6,textvariable=self.totald,width=10,font= ("Times New Roman",16,"bold"),bd=5,relief="sunken").grid(row=2,column=1,padx=10,pady=1)
        
        #============label for taxes================
        gsa_lbl = Label(F6,text="Geometry GST",font=("Times New Roman",14,'bold'),bg="white").grid(row=0, column=2,padx=20,pady=1,sticky="w")
        gsa_txt = Entry(F6,textvariable=self.gstsa,width=10,font= ("Times New Roman",16,"bold"),bd=5,relief="sunken").grid(row=0,column=3,padx=10,pady=1)

        gs_lbl = Label(F6,text="Journals GST",font=("Times New Roman",14,'bold'),bg="white").grid(row=1, column=2,padx=20,pady=1,sticky="w")
        gs_txt = Entry(F6,textvariable=self.gsts,width=10,font= ("Times New Roman",16,"bold"),bd=5,relief="sunken").grid(row=1,column=3,padx=10,pady=1)
        
        gd_lbl = Label(F6,text="Painting Items GST",font=("Times New Roman",14,'bold'),bg="white").grid(row=2, column=2,padx=20,pady=1,sticky="w")
        gd_txt = Entry(F6,textvariable=self.gstd,width=10,font= ("Times New Roman",16,"bold"),bd=5,relief="sunken").grid(row=2,column=3,padx=10,pady=1)

       
        #============frame for calculation buttons=========
        btn_F = Frame(F6,bd=6,relief="groove")
        btn_F.place(x=750,width=580,height=105)
        
        total_btn = Button(btn_F,command=self.total,text="Total",pady=12,width=10,font="arial 15 bold",bd=4).grid(row=0,column=0,padx=5,pady=5)
        gbill_btn = Button(btn_F,command=self.Receipt,text="Generate Bill",pady=12,width=10,font="arial 15 bold",bd=4).grid(row=0,column=1,padx=5,pady=5)
        clear_btn = Button(btn_F,command=self.clear_data,text="Clear",pady=12,width=10,font="arial 15 bold",bd=4).grid(row=0,column=2,padx=5,pady=5)
        exit_btn = Button(btn_F,command=self.exit_system,text="Exit",pady=12,width=9,font="arial 15 bold",bd=4).grid(row=0,column=3,padx=5,pady=5)
        self.receipt_details()
        
        
    def total(self):
        #for geometry
        self.asan = self.a.get()*420
        self.scalesan = self.scale.get()*400
        self.csan = self.c.get()*390
        self.dsan = self.d.get()*500
                
        self.total_sa_price=float(
                                        self.asan+
                                        self.scalesan+
                                        self.csan+
                                        self.dsan
                                        )        
        
        self.tg.set(str(self.total_sa_price))
        self.sa_tax=round((self.total_sa_price*0.13),2) #variable for gerometry gst
        self.gstsa.set("rs."+str(self.sa_tax))
        
  
        #for Journal      
        self.hsal = self.h.get()*420
        self.jsal = self.j.get()*400
        self.ksal = self.k.get()*390
        self.lsal = self.l.get()*500
                
        self.total_s_price=float(
                                        self.hsal+
                                        self.jsal+
                                        self.ksal+
                                        self.lsal
                                        )        
        
        self.totals.set(str(self.total_s_price))
        self.s_tax=round((self.total_s_price*0.13),2) #variable for geometry gst
        self.gsts.set("rs."+str(self.s_tax))
        
       
        #for Painting Items
        self.zd = self.z.get()*400
        self.xd = self.x.get()*380
        self.cad = self.ca.get()*300
        self.vd = self.v.get()*490
        
        self.total_d_price=float(
                                           self.zd+
                                           self.xd+
                                           self.cad+
                                           self.vd
                                           )        
        
        self.totald.set(str(self.total_d_price))
        self.d_tax=round((self.total_d_price*0.13),2)#variable for geometry gst
        self.gstd.set("rs."+str(self.d_tax))
        
        self.bill_total =float( self.total_sa_price + self.total_s_price + self.total_d_price+
                                self.sa_tax + self.s_tax + self.d_tax
                                 )
        
        
    def receipt_details(self):  
        self.txtarea.delete("1.0",END)
        self.txtarea.insert(END,"\t**Welcome to Stationary Shop**")
        self.txtarea.insert(END,f"\n\n Bill No: {self.cbill.get()}")
        self.txtarea.insert(END,f"\n Customer Name: {self.cname.get()}")
        self.txtarea.insert(END,f"\n Phone No: {self.cphone.get()}")
        self.txtarea.insert(END,f"\n=======================================")
        self.txtarea.insert(END,f"\n Product\t\t\tQty\tPrice")
        self.txtarea.insert(END,f"\n=======================================")
        

        
    def Receipt(self):
        if self.cname.get()=="" or self.cphone.get()=="":
            messagebox.showerror("Error","please enter customer details first")
        elif self.tg.get()=="0.0" and self.totals.get()=="0.0" and self.totald.get()=="0.0":
            messagebox.showerror("Error","please select some products")
        else:
            self.receipt_details()
            #geometry
            if self.a.get()!=0:
                self.txtarea.insert(END,f"\n Pencil          \t\t\t{self.a.get()}\t{self.asan}")
                
            if self.scale.get()!=0:
                self.txtarea.insert(END,f"\n Scale      \t\t\t{self.scale.get()}\t{self.scalesan}")
                
            if self.c.get()!=0:
                self.txtarea.insert(END,f"\n Eraser        \t\t\t{self.c.get()}\t{self.csan}")
                
            if self.d.get()!=0:
                self.txtarea.insert(END,f"\n Sharpener       \t\t\t{self.d.get()}\t{self.dsan}")
                
                
    
                
            #Journal
            if self.h.get()!=0:
                self.txtarea.insert(END,f"\n 200 Pages    \t\t\t{self.h.get()}\t{self.hsal}")
                
            if self.j.get()!=0:
                self.txtarea.insert(END,f"\n 300 Pages\t\t\t{self.j.get()}\t{self.jsal}")
                
            if self.k.get()!=0:
                self.txtarea.insert(END,f"\n 400 Pages  \t\t\t{self.k.get()}\t{self.ksal}")
                
            if self.l.get()!=0:
                self.txtarea.insert(END,f"\n 500 Pages \t\t\t{self.l.get()}\t{self.lsal}")
                
                
             
             #Painting Items
            if self.z.get()!=0:
                self.txtarea.insert(END,f"\n Oil Paint                 \t\t\t{self.z.get()}\t{self.zd}")
                
            if self.x.get()!=0:
                self.txtarea.insert(END,f"\n Oil Pastels                  \t\t\t{self.x.get()}\t{self.xd}")
                
            if self.ca.get()!=0:
                self.txtarea.insert(END,f"\n Crayons               \t\t\t{self.ca.get()}\t{self.cad}")
                
            if self.v.get()!=0:
                self.txtarea.insert(END,f"\n Painting Sheets              \t\t\t{self.v.get()}\t{self.vd}")
                
                
            
            self.txtarea.insert(END,f"\n---------------------------------------")
            if self.gstsa.get()!="rs.0.0":
                self.txtarea.insert(END,f"\n Geometry gst              \t\t\t{self.gstsa.get()}")
    
            if self.gsts.get()!="rs.0.0":
                self.txtarea.insert(END,f"\n Journal gst                 \t\t\t{self.gsts.get()}")
           
            if self.gstd.get()!="rs.0.0":
                self.txtarea.insert(END,f"\n Painting Items gst           \t\t\t{self.gstd.get()}")
            
            self.txtarea.insert(END,f"\n Total Bill:            \t\t\t Rs. {str(self.bill_total)}")
            self.txtarea.insert(END,f"\n---------------------------------------")
            self.save_bill() 
    
    def save_bill(self):
        sb = messagebox.askyesno("Save Bill", "Do you want to save this bill?")
        if sb>0:
            self.bill_data=self.txtarea.get("1.0",END)
            f1 = open("bills/"+str(self.cbill.get())+".txt","w")
            f1.write(self.bill_data)
            f1.close()
            messagebox.showinfo("Saved",f"bill no.{self.cbill.get()} has been saved successfully")
        else:
            return
        
    def find_bill(self):
        show="no"
        for i in os.listdir("bills/"):
            if i.split('.')[0]==self.search_bill.get():
                f1 = open(f"bills/{i}","r")
                self.txtarea.delete('1.0',END)
                for d in f1:
                    self.txtarea.insert(END,d)
                f1.close()
                show="yes"
        if show=="no":
            messagebox.showerror("Error","Invalid Bill Number")
            
    def clear_data(self):
         sb=messagebox.askyesno("Exit","Do you want to clear the data?")
         if sb>0:
           
            #==================Geometry variable=================
            self.a.set(0)
            self.scale.set(0)
            self.c.set(0)
            self.d.set(0)
            
            #==================Journal variable====================
            self.h.set(0)
            self.j.set(0)
            self.k.set(0)
            self.l.set(0)
            
            #==================Painting Items variable===============
            self.z.set(0)
            self.x.set(0)
            self.ca.set(0)
            self.v.set(0)
            
            
            #==================total variable====================
            self.tg.set("")
            self.totals.set("")
            self.totald.set("")
            
            #==================taxes variable====================
            self.gstsa.set("")
            self.gsts.set("")
            self.gstd.set("")
            
            #==================calculation variable===============
            self.cname.set("")
            self.cphone.set("")
            
            self.cbill.set("")
            x=random.randint(1000,9999)
            self.cbill.set(str(x))
    
            self.search_bill.set("")
            self.receipt_details()
                

    def exit_system(self):
        sb=messagebox.askyesno("Exit","Do you want to exit from Inventory system?")
        if sb>0:
            self.screen.destroy() 
            
            
screen = Tk()  
S = InventorySystem(screen)
screen.mainloop() 
```
